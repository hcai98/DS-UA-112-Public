### Stem and Leaf Plot

Stem and Leaf plots display numerical data. Numbers are gathered together according to common digits. Create a stem and leaf plot for the following data 

44, 46, 47, 49, 63, 64, 66, 68, 68, 72, 72, 75, 76, 81, 84, 88, 106

```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
df = pd.DataFrame({"data": [44, 46, 47, 49, 63, 64, 66, 68, 68, 72, 72, 75, 76, 81, 84, 88, 106]})
```

```python
df['first_digits'] = df['data'].apply(lambda x: int(str(x)[:-1]))
df['last_digit'] = df['data'].apply(lambda x: int(str(x)[-1]))
```

```python
df
```

```python
grouped = df.groupby(df['first_digits'])

for name, value in grouped:
    print(name)
    display(value)
```

```python
grouped = df.groupby(df['first_digits'])

for name, value in grouped:
    print(value['first_digits'].values[0], "|", value['last_digit'].values) 
```
